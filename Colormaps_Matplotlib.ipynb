```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

```python
cmaps = sorted(plt.colormaps())
print(cmaps) #all colormaps in matplotlib
```

```python
colormap_info = {} # dictionary -> colormap: unique colors

for col_map in cmaps:
    cmap = plt.get_cmap(col_map, 512)
    # print(cmap(0))
    rgba_colors = cmap(np.linspace(0, 1, 512)) # get RGB + alpha value for each colormap
    unique_colors = set(tuple(rgba) for rgba in rgba_colors) # remove duplicate colors
    colormap_info[col_map] = len(unique_colors) # number of unique color

i = 0for col_map, num_colors in colormap_info.items():
    if i == 8: break    print(f"Colormap: {col_map}, Number of Shades: {num_colors}")# colormaps with ~512 colors have that many shades available    i += 1
```

```python
num_colors = 256
num_rows = len(cmaps)

fig, axes = plt.subplots(num_rows, 1, figsize=(8, 2*num_rows))

for i, col_map in enumerate(cmaps):
    cmap = plt.get_cmap(col_map, num_colors)
    values = np.linspace(0, 1, num_colors).reshape(1,-1)
    
    ax = axes[i]
    ax.imshow(values, aspect='auto', cmap=cmap) # simply plot the cmap value using imshow
    
    ax.set_title(col_map)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0, num_colors - 1)
    ax.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```
